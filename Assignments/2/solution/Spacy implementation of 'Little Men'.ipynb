# Importing spacy and running the book 'Little Men'

```python
import spacy
nlp = spacy.load("en_core_web_sm")
test_file = open("/Users/sumeetmishra/desktop/ThirdSem/ANLP/Assignments/2/Little Men, by Louisa May Alcott.txt", "r", encoding = "utf8")
text = test_file.read()

#desired = []
doc = nlp(text)

for sentence in doc.sents:
    sent = nlp(sentence.text)
    for token in sent:
        print(token.text, token.lemma_, token.pos_, token.tag_, token.dep_)
```
