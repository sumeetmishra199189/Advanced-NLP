**c. Calculate the Variance of this code, and by code I mean the distribution of the point-wise entropy. :-)**

```
from math import log

def entropy(p):
    res = 0.0
    for x in p:
        res += x * log(x, 2)
    return -res
```

```
print(entropy(relfrq))
```

```
entdist = [ -x * log(x, 2) for x in relfrq ]
j=0
for i in myFD:
  print('Entropy for ',i,':',entdist[j])
  j+=1
```

```
import numpy as np
print('Variance of the pointwise entropy distribution is: ', np.var(entdist))
```
